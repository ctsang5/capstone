## Clean the Data

```python
import pandas as pd
from pathlib import Path

import numpy as np
```

```python
options_path = "../data/options_data.csv"
options = pd.read_csv(options_path)
```

```python
options = options.iloc[:, :7]
```

```python
options
```

```python
options = options.dropna()
```

```python
options
```

```python
options = options.rename(columns = {'equity':'symbol', 'volatil':'iv', 'last':'last_price', 'expire':'expiry', 'indicator':'type', 'underlying':'underlying_price'})
```

```python
options
```

```python
options.to_csv('../data/options_data.csv', index= False)
```

```python
conditions = [
    (options['type'] == 'Call') & (options['strike'] < options['underlying_price']),  # Call ITM
    (options['type'] == 'Call') & (options['strike'] > options['underlying_price']),  # Call OTM

    (options['type'] == 'Put') & (options['strike'] > options['underlying_price']),   # Put ITM
    (options['type'] == 'Put') & (options['strike'] < options['underlying_price'])    # Put OTM
]

choices = ['ITM', 'OTM', 'ITM', 'OTM']

options['moneyyyyy'] = np.select(conditions, choices)

print(pd.crosstab(options['type'], options['moneyyyyy']))
```
